```python
import requests, json
import csv
import pandas as pd
import numpy as np
```

```python
url = "https://cycling.data.tfl.gov.uk/CyclingInfrastructure/data/lines/cycle_lane_track.json"


content = requests.get(url)
json = json.loads(content.content)
```

```python
f = open("cycle_lane.csv", "w")
f.truncate()
f.close()
```

```python
cycle_lane = json['features'] 

data_file = open('cycle_lane.csv', 'w') 

csv_writer = csv.writer(data_file) 
```

```python
count = 0
  
for i in cycle_lane:
    if count == 0: 
  
        # Writing headers of CSV file 
        header = i['properties'].keys() 
        csv_writer.writerow(header) 
        count += 1
  
    # Writing data of CSV file 
    csv_writer.writerow(i['properties'].values()) 
    
data_file.close()
```

```python
year_2017=df2.loc[(df2['year'] == 2017)]
year_2018=df2.loc[(df2['year'] == 2018)]
year_2019=df2.loc[(df2['year'] == 2019)]
```

```python
df_2= year_2018.merge(year_2017, on='BOROUGH',how='left')
```

```python
df3= df_2.merge(year_2019, on='BOROUGH',how='left')
df3
```

```python
df3.rename({'year_x':'y_2018','year_y':'y_2017','year':'y_2019',},axis='columns',inplace=True)

df3
```

```python
#df3= df3.fillna(value=0)
#df['DataFrame Column'] = df['DataFrame Column'].replace(np.nan, 0)
df3 = df3.replace(np.nan, 0)
df3
```

```python
year_2017_=df3[['y_2017','infrastructure_y']]
year_2018_=df3[['y_2018','infrastructure_x']]
year_2019_=df3[['y_2019','infrastructure']]
```

```python
year_2017_n= year_2017_['infrastructure_y'].values.tolist()
year_2018_n= year_2018_['infrastructure_x'].values.tolist()
year_2019_n= year_2019_['infrastructure'].values.tolist()
```

```python
borough=df3.BOROUGH.values.tolist()
```

```python
import matplotlib.pyplot as plt
import numpy as np


labels = borough
men_means = year_2017_n
no_means=year_2018_n
women_means = year_2019_n

x = np.arange(len(labels))  # the label locations
width = 0.37  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/3, men_means, width, label='2017')
rects2 = ax.bar(x + width/3, no_means, width, label='2018')
rects3 = ax.bar(x + 3*width/3, women_means, width, label='2019')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('number of infrastructure',fontsize=17)
ax.set_title('Changes by borough (2017-2019)')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=90,fontsize=13)
#ax.set_xticklabels(xlabels, rotation=40, ha=ha[n])
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


#autolabel(rects1)
#autolabel(rects2)

fig.tight_layout()
fig.set_size_inches(18.5, 20.5)
fig.savefig('infras.png', dpi=100)

plt.show()
```
