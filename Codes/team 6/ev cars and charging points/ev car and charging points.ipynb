```python
import pandas as pd
```

```python
charging=pd.read_csv('charge point1.csv')
population=pd.read_csv('population.csv')
Car=pd.read_csv('DATA.csv')
```

```python
charging
```

```python
population
```

```python
All=pd.merge(charging,population,left_on='Borough',right_on='Area name',how='left')
All
```

```python
All1=pd.merge(All,Car,left_on='Borough',right_on='Local Authority  ',how='left')
```

```python
All1
```

```python
All1['charg_percent']=All1['Charging point']*1000/(All1['Cars']*All1['2015'])
```

```python
All1['charg_percent']
```

```python
import numpy as np
X=[]
arr = np.asarray(All1['charg_percent'])
for x in arr:
    x = float(x - np.min(arr))/(np.max(arr)- np.min(arr))
    X.append(x)
All1['Nomalize_charge']=X
All1
```

```python
All1['ev_percent']=All1['ev']*1000/(All1['Total']*All1['2015'])
All1
```

```python
import numpy as np
B=[]
arr = np.asarray(All1['ev_percent'])
for x in arr:
    x = float(x - np.min(arr))/(np.max(arr)- np.min(arr))
    B.append(x)
All1['Nomalize_evcars']=B
All1
```

```python
All1.to_csv('All.csv',index=False)
```
