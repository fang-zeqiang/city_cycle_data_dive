```python
import pandas as pd
```

```python
data=pd.read_csv('LondonCycling2019sum.csv')
```

```python
data
```

```python
#change the type of Date
pd.to_datetime(data.Date,format='%Y/%m/%d')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,10))

ax1 = plt.subplot(211)
plt.plot(data.Date,data.Count)
plt.ylabel("The times of Cycling") 
plt.title("The trend of cycling usage times in London, 2019")

ax2 = plt.subplot(212)
plt.plot(data.Date,data.AVG_Duration)
plt.ylabel("The average duration of Cycling") 
plt.title("The trend of cycling usage duration in London, 2019")

plt.savefig('The trend of variables change.png',dpi=300)
```

The distribution of counts and durations

```python
plt.figure(figsize=(10,5))

ax1=plt.subplot(121)
data.Count.hist(bins=20)
plt.title('Hitogram of Counts')
plt.xlabel('Counts of Cycling')
plt.ylabel('Frequency')

ax2=plt.subplot(122)
data.AVG_Duration.hist(bins=20)
plt.title('Hitogram of Durations')
plt.xlabel('Durations of Cycling')
plt.ylabel('Frequency')

plt.savefig('The distribution of Counts and Durations',dpi=300)
```

The cycling usage in different seasons

```python
# Average count and duration of cycling usage in every seasons

se_count=[]
se_duration=[]
seasons=['Q1','Q2','Q3','Q4']

for i in seasons:
    a=data[data['Quarter']==i].Count.mean()
    b=data[data['Quarter']==i].AVG_Duration.mean()
    se_count.append(a)
    se_duration.append(b)
    
plt.figure(figsize=(12,5))

ax1=plt.subplot(121)
plt.bar(['Spring','Summer','Fall','Winter'],se_count)
plt.ylabel('The average number of cycling usage')
plt.title('The usage of cycling in different seasons (Counts) ')

ax2=plt.subplot(122)
plt.bar(['Spring','Summer','Fall','Winter'],se_duration)
plt.ylabel('The average duration of cycling usage')
plt.title('The usage of cycling in different seasons (Durations) ')

plt.savefig('The cycling usage in different seasons',dpi=300)
```

The cycling usage in weekdays and weekends

```python
week_count=[]
week_duration=[]
week=['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
week_label=['Monday','Tuesday','Wednesday','Thuesday','Friday','Saturday','Sunday']

for i in week:
    a=data[data['Week']==i].Count.mean()
    b=data[data['Week']==i].AVG_Duration.mean()
    week_count.append(a)
    week_duration.append(b)

plt.figure(figsize=(18,5))

ax1=plt.subplot(121)
plt.bar(week_label,week_count)
plt.ylabel('The average number of cycling usage')
plt.title('The usage of cycling in different days (Counts) ')

ax2=plt.subplot(122)
plt.bar(week_label,week_duration)
plt.ylabel('The average duration of cycling usage')
plt.title('The usage of cycling in different days (Durations) ')

plt.savefig('The cycling usage in weekdays and weekends',dpi=300)
```

The cycling usage in different temperature ranges

```python
# cheak the range of temperature

print(f'The max and min:{data["Avg_Temperature"].max()} {data["Avg_Temperature"].min()}')
print(f'The range: {data["Avg_Temperature"].max()-data["Avg_Temperature"].min()}')
```

```python
# break the temperature into 7 blocks
temp=[28.3,36.3,44.3,52.3,60.3,68.3,76.3,85]
temp_label=['28.3~36.3','36.3~44.3','44.3~52.3','52.3~60.3','60.3~68.3','68.3~76.3','76.3~84.3']
```

```python
temp_count=[]
temp_duration=[]


for i in range(0,7):
    a=data.loc[(data.Avg_Temperature>=temp[i])&(data.Avg_Temperature<temp[i+1])].Count.mean()
    b=data.loc[(data.Avg_Temperature>=temp[i])&(data.Avg_Temperature<temp[i+1])].AVG_Duration.mean()
    temp_count.append(a)
    temp_duration.append(b)

plt.figure(figsize=(18,5))

ax1=plt.subplot(121)
plt.bar(temp_label,temp_count)
plt.ylabel('The average number of cycling usage')
plt.xlabel('Temperature (°F)')
plt.title('The usage of cycling in different temperature(Counts) ')

ax2=plt.subplot(122)
plt.bar(temp_label,temp_duration)
plt.xlabel('Temperature (°F)')
plt.ylabel('The average duration of cycling usage')
plt.title('The usage of cycling in different temperature(Durations) ')

plt.savefig('The cycling usage in different temperatures',dpi=300)
```
