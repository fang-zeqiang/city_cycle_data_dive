```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read CO2 emission data
E_CO=pd.read_csv("data/emission_CO2.csv")
```

```python
E_CO.rename({'Row Labels':'NAME'},axis=1, inplace=True)
```

```python
E_CO
```

```python
#borough data
borough = gpd.read_file('data/ESRI/London_Borough_Excluding_MHW.shp')
borough
```

```python
#
Join = borough.merge(E_CO, on="NAME")
```

```python
Join 
```

```python
Join.plot(column='Total Road Transport', scheme='equal_interval', k=10, 
         cmap=plt.cm.Reds_r, alpha=1, 
         edgecolor='w', linewidth=0.1)
```

```python
E_CO_New=E_CO.copy()
```

```python
E_CO_New[' Petrol Car'] =E_CO[' Petrol Car']-0.05*E_CO[' Petrol Car'].values*0.958 #reduce 0.05car ,emission 0.042
E_CO_New[' Diesel Car'] =E_CO[' Diesel Car']-0.05*E_CO[' Diesel Car']*0.958
#ECO['Electric Car'] =0.95*ECO['Electric Car']
E_CO_New[' TfL Bus']=E_CO[' TfL Bus']-0.25*E_CO[' TfL Bus']*0.985
E_CO_New[' Non-TfL Bus and Coach']=E_CO[' Non-TfL Bus and Coach']-0.25*E_CO[' Non-TfL Bus and Coach']*0.985
```

```python
E_CO_New
```

```python
#sum emission
E_CO_New['Total Road Transport']=E_CO_New.iloc[:,1:13].sum(axis=1)
```

```python
E_CO_New
```

```python
#afer e-bike
Join_n = borough.merge(E_CO_New, on="NAME")
```

```python
E_CO_New['differ']=E_CO['Total Road Transport']-E_CO_New['Total Road Transport']
```

```python
Join.dtypes
```

```python
Join
```

```python
f, ax = plt.subplots(1)
f.suptitle("CO2 Emission in London(Tonnes/Year)")
Join.plot(ax=ax,column='Total Road Transport',legend=True,
         cmap=plt.cm.Reds, alpha=1, 
         edgecolor='w', linewidth=0.1)
ax.set_axis_off()
plt.show()
```

```python
f, ax = plt.subplots(1)
f.suptitle("Reduced CO2 Emission (Tonnes/Year) after the Usage of Micro-mobility ")
Join_n.plot(ax=ax,column='differ', legend=True, 
         cmap="GnBu", alpha=1, 
         edgecolor='w', linewidth=0.1)
ax.set_axis_off()
plt.show()
 #scheme='equal_interval'
```
